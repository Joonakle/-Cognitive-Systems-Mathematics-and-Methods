```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(1, 11)
x1 = np.power(x, 3)
x2 = np.absolute(x)
x2a = x+0.00000001
x3 = np.log(x2a)
x4 = np.sqrt((100+x))
x5 = (0.01*x*np.power(2.71828, x))
y = (x1+x2+x3)/(x4+x5)

plt.plot(x,y)
plt.show()
```
